# **Dealing with missing data**

1, Drop missing values

2, Fill in missing values

a. Mean
                
                mean = df['column'].mean()

-> Replace missing value with mean value
                
                df['column'].fillna(mean, inplace = True)

b. Preceding, following values



```python
#Calculate mean value, fill with mean value
mean = round(df1["avg_salary"].mean(), 2)
df1["avg_salary"].fillna(mean, inplace=True)
```

```python
#Fill missing values with bfill
df1["max_salary"].bfill(inplace=True)
```

```python
#Fill missing values with ffill
df1["min_salary"].ffill(inplace=True)
```

```python
df1["avg_salary"] = (df1["min_salary"] + df1["max_salary"])/2
```

```python
#Checking for missing value again
df1
```
